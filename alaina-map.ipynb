```python
import folium
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/300k.csv')
```

```python
data["appearedLocalTime"] = pd.to_datetime(data["appearedLocalTime"])
data["hourAppeared"] = data["appearedLocalTime"].map(lambda x: x.hour)
```

```python
locations = data[["pokemonId", "latitude", "longitude", "hourAppeared", "continent"]]
```

```python
location_subset = locations[["pokemonId", "latitude", "longitude", "hourAppeared"]][locations["continent"] == "America"]
location_subset = location_subset[["pokemonId", "latitude", "longitude", "hourAppeared"]][(location_subset["latitude"] < 47.8) & (location_subset["latitude"] > 47.4)]
```

```python
location_subset
```

```python
folium_map = folium.Map(location=[47.6062, -122.3321],
                        zoom_start=12,
                        tiles="CartoDB dark_matter")
marker = folium.CircleMarker(location=[40.738, -73.98])
marker.add_to(folium_map)

# popup_text = "pokemon ID: " + str(row["pokemonId"]) + "<br> hour appeared: " + str(row["hourAppeared"]) + "<br>"
```

```python
for index, row in location_subset.iterrows():
    time = row["hourAppeared"]
    if time > 6 and time < 20 :
        color="#E37222"
    else :
        color="#0A8A9F"
    popup = folium.Popup(row["pokemonId"], parse_html=True)
    folium.CircleMarker(location=(row["latitude"],
                                  row["longitude"]),
                        radius=2,
                        color=color,
                        popup=popup,
                        fill=True,
                        fill_color=color).add_to(folium_map)
```
