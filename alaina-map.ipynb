```python
import folium
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/300k.csv')
```

```python
data["appearedLocalTime"] = pd.to_datetime(data["appearedLocalTime"])
data["hourAppeared"] = data["appearedLocalTime"].map(lambda x: x.hour)
```

```python
locations = data.groupby(['latitude', 'longitude']).first().reset_index()
asia_subset = locations[["pokemonId", "latitude", "longitude", "hourAppeared"]][locations["continent"] == "Asia"]
japan_subset = asia_subset[["pokemonId", "latitude", "longitude", "hourAppeared"]][(asia_subset["latitude"] < 36) & (asia_subset["latitude"] > 34)]
```

```python
america_subset = locations[["pokemonId", "latitude", "longitude", "hourAppeared"]][locations["continent"] == "America"]
```

```python
losangeles_subset = america_subset[["pokemonId", "latitude", "longitude", "hourAppeared"]][(america_subset["longitude"] < -118.15) & (america_subset["longitude"] > -118.67)]
```

```python
denver_subset = america_subset[["latitude", "longitude", "hourAppeared"]][(america_subset["longitude"] > -105.05) & (america_subset["longitude"] < -104.8)]
denver_subset = denver_subset[(denver_subset["latitude"] > 39.6) & (denver_subset["latitude"] < 39.8)]
```

```python
chicago_subset = america_subset[["latitude", "longitude", "hourAppeared"]][(america_subset["longitude"] > -87.86) & (america_subset["longitude"] < -87.52)]
chicago_subset = chicago_subset[(chicago_subset["latitude"] > 41.6) & (chicago_subset["latitude"] < 42.02)]
```

```python
ny_subset = america_subset[["latitude", "longitude", "hourAppeared"]][(america_subset["longitude"] > -74.04) & (america_subset["longitude"] < -73.7)]
ny_subset = ny_subset[(ny_subset["latitude"] > 40.5) & (ny_subset["latitude"] < 40.9)]
```

```python
folium_map = folium.Map(location=[40.6, -73.87],
                        zoom_start=13,
                        tiles="CartoDB dark_matter")

# popup_text = "pokemon ID: " + str(row["pokemonId"]) + "<br> hour appeared: " + str(row["hourAppeared"]) + "<br>"
```

```python
for index, row in ny_subset.iterrows():
    time = row["hourAppeared"]
    if time > 6 and time < 20 :
        color="#E37222"
    else :
        color="#0A8A9F"
    #popup = folium.Popup(row["pokemonId"], parse_html=True)
    
    folium.Circle(location=(row["latitude"],
                                  row["longitude"]),
                        radius=2,
                        color=color,
                        #popup=popup,
                        fill=True,
                        fill_color=color).add_to(folium_map)
```
